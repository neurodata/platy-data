```python
left = pymaid.get_skids_by_annotation("left")
right = pymaid.get_skids_by_annotation("right")
hemis = {"left": len(left), "right": len(right)}
plt.bar(hemis.keys(), hemis.values()) 
plt.title("# of Neurons in each Hemisphere")
plt.xlabel("Hemisphere")
plt.ylabel("# of Neurons")
```

```python
sensory = pymaid.get_skids_by_annotation("Sensory neuron")
inter = pymaid.get_skids_by_annotation("interneuron")
motor = pymaid.get_skids_by_annotation("motorneuron")
types = {"sensory": len(sensory), "inter": len(inter), "motor": len(motor)}
plt.bar(types.keys(), types.values()) 
plt.title("# of Neurons for Different Types")
plt.xlabel("Type")
plt.ylabel("# of Neurons")
```

```python
segment0 = pymaid.get_skids_by_annotation("segment_0")
segment1 = pymaid.get_skids_by_annotation("segment_1")
segment2 = pymaid.get_skids_by_annotation("segment_2")
segment3 = pymaid.get_skids_by_annotation("segment_3")
segments = {"0": len(segment0), "1": len(segment1), "2": len(segment2), "3": len(segment3)}
plt.bar(segments.keys(), segments.values()) 
plt.title("# of Neurons in Each Segment")
plt.xlabel("Segment")
plt.ylabel("# of Neurons")
```

```python
celltypes = []
num_neurons = []
celltypes_count = {}
for i, digit in enumerate(range(1, 181)):
   celltypes.append("celltype{}".format(str(digit)))
   num_neurons.append(len(pymaid.get_skids_by_annotation(celltypes[i])))
   celltypes_count[celltypes[i]] = [num_neurons[i]]

df = pd.DataFrame(celltypes_count, index=["# of neurons"])
print(df)
```
